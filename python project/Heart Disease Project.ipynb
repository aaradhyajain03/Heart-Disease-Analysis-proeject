```python
#Data Cleaning
df.dropna(inplace=True)  # Drop rows with missing values
df.drop_duplicates(inplace=True) #Drop duplicate value
```

```python
#Correlation Matrix
corr_matrix = df.corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
#count the unique value
target_counts =df['target'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(target_counts.index, target_counts.values, color=['skyblue', 'salmon'])
plt.title('Count of Target Values')
plt.xlabel('Target')
plt.ylabel('Count')
plt.xticks(target_counts.index, ['No Heart Disease', 'Heart Disease']) 
plt.show()
```

```python
sex_counts = df['sex'].value_counts()
```

```python
sex_counts = df['sex'].value_counts()

# Plot bar chart using Matplotlib
plt.figure(figsize=(8, 6))
plt.bar(sex_counts.index, sex_counts.values, color=['lightblue', 'lightcoral'])
plt.title('Count of Sex Values')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(sex_counts.index, ['Male', 'Female']) 
plt.show()
```

```python
#gender distribution acc to target variable
grouped_data = df.groupby(['target', 'sex']).size().unstack()

grouped_data.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'lightcoral'])
plt.title('Gender Distribution According to Target Variable')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.legend(['Female', 'Male'], title='Sex')
plt.show()
```

```python
#age distribution 
plt.figure(figsize=(10, 6))

plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')

plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')

plt.show()
```

```python
#check chest pain type
```

```python
cp_counts = df['cp'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(cp_counts.index, cp_counts.values, color='lightgreen', edgecolor='black')
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Patients')
plt.xticks(cp_counts.index, ['Typical Angina', 'Atypical Angina', 'Non-Anginal Pain', 'Asymptomatic'])

plt.show()
```

```python
#show the chest apin distribution as per target variable
```

```python
grouped_cp = df.groupby(['cp', 'target']).size().unstack()

grouped_cp.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'salmon'], edgecolor='black')

plt.title('Chest Pain Type Distribution as per Target Variable')
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Patients')
plt.xticks(cp_counts.index, ['Typical Angina', 'Atypical Angina', 'Non-Anginal Pain', 'Asymptomatic'],rotation=0)
plt.legend(['No Heart Disease', 'Heart Disease'], title='Target')
plt.show()
```

```python
#show fasting blood suagr distribution as per target variable
```

```python
grouped_fbs = df.groupby(['fbs', 'target']).size().unstack()

grouped_fbs.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'salmon'], edgecolor='black')

plt.title('Fasting Blood Sugar Distribution as per Target Variable')
plt.xlabel('Fasting Blood Sugar (1 = FBS > 120 mg/dl, 0 = FBS â‰¤ 120 mg/dl)')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.legend(['No Heart Disease', 'Heart Disease'], title='Target')
plt.show()
```

```python
#check resting blood pressure distribution

plt.figure(figsize=(10, 6))

plt.hist(df['trestbps'], bins=20, color='lightblue', edgecolor='black')

plt.title('Resting Blood Pressure Distribution')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Number of Patients')
plt.show()
```

```python
#compare resting blood pressure as per sex column
```

```python
plt.figure(figsize=(10, 6))


df.boxplot(column='trestbps', by='sex', grid=False, patch_artist=True, 
           boxprops=dict(facecolor='lightblue', color='black'),
           whiskerprops=dict(color='black'),
           capprops=dict(color='black'),
           medianprops=dict(color='red'))

plt.title('Resting Blood Pressure Distribution by Gender')
plt.suptitle('')  
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Resting Blood Pressure (mm Hg)')

# Show the plot
plt.show()
```
